```python
import matplotlib.pyplot as plt
import numpy as np
import librosa
import librosa.display

# Assuming 'dataset' is loaded and has an 'audio' column with 'array' and 'sampling_rate'
# Let's pick a random sample for waveform visualization
sample_audio = dataset[0]['audio']
audio_array = sample_audio['array']
sampling_rate = sample_audio['sampling_rate']

plt.figure(figsize=(14, 5))
librosa.display.waveshow(y=audio_array, sr=sampling_rate)
plt.title('Waveform of a Sample Audio')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```

```python
# Calculate audio lengths
audio_lengths = []
for i in range(len(dataset)):
    sample_audio = dataset[i]['audio']
    audio_array = sample_audio['array']
    sampling_rate = sample_audio['sampling_rate']
    length_seconds = len(audio_array) / sampling_rate
    audio_lengths.append(length_seconds)

# Plot histogram of audio lengths
plt.figure(figsize=(10, 6))
plt.hist(audio_lengths, bins=50, edgecolor='black')
plt.title('Distribution of Audio Lengths')
plt.xlabel('Audio Length (seconds)')
plt.ylabel('Number of Samples')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```
